```python
import pandas as pd
import numpy as np
```

```python
def generate_varspace(guess, variation=0.2, N=5):
    var_space = np.linspace(guess*(1-variation), guess*(1+variation), N)
    
    if type(guess) == int:
        var_space = [int(round(var_space[i],0)) for i in range(len(var_space))]
    else:
        var_space = [round(var_space[i],6) for i in range(len(var_space))]
    return var_space
```

```python
# to completely overwrite the file:
#df = pd.read_csv('parameters-template.csv')
# to add onto the existing file:
df = pd.read_csv('parameters.csv')
```

```python
Group_name = 'MaxRange'
group_index = 0

CoefL        = [0.000114] 
#CoefL        = generate_varspace(0.000114,0.5,11)
CoefT        = [0.00037]  
#CoefT        = generate_varspace(0.00037,0.5,11)
Gain         = [10000]
#Gain         = generate_varspace(10000,0.5,11)
GETGain      = [1000] 
#GETGain      = generate_varspace(1000,0.5,11)
PeakingTime  = [720] 
#PeakingTime  = generate_varspace(720,0.5,11)
#MaxRange     = [500] 
MaxRange     = generate_varspace(500,0.5,11)

for CL in CoefL:
    for CT in CoefT:
        for Gn in Gain:
            for GG in GETGain:
                for PT in PeakingTime:
                    for MR in MaxRange:
                        df.loc[len(df.index)] = [Group_name+str(group_index),
                                                CL, CT, Gn, GG, PT, MR, 0]
                        group_index += 1
df.to_csv('parameters.csv',index=False)
```
