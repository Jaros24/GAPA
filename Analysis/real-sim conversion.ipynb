```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
import h5py
import os, sys
```

```python
output_directory = "C:\\Users\\Adam\\OneDrive - Tenundra, Inc\\Desktop\\Programming\\GADGET2\\simOutput\\"
target_h5 = '273_800p'


# clone the h5 file from the original
f1 = h5py.File("C:\\Users\\Adam\\OneDrive - Tenundra, Inc\\Desktop\\273_800p.h5", "r")
f2 = h5py.File(output_directory + target_h5 + '.h5', "w")

for key in f1.keys():
    f2.copy(f1[key], key)

f1.close()
f2.close()
```

```python
# Read in the data
f = h5py.File(output_directory + target_h5 + '.h5', 'r+')
print(f.keys())
print(f['erace'].keys())
print(f['event'].keys())

# Move data from erace and event to top of h5
for key in f['erace'].keys():
    f[key] = f['erace'][key]
for key in f['event'].keys():
    f[key] = f['event'][key]
del f['erace']
del f['event']

# reindex the data
trace_indexes = []
for key in f.keys():
    if 'Trace_' in key:
        ind = int(key.split('_')[1][1:-1])
        trace_indexes.append(ind)

for i in range(len(trace_indexes)):
    f['Trace_[' + str(i) + ']'] = f['Trace_[' + str(trace_indexes[i]) + ']' ]
    del f['Trace_[' + str(trace_indexes[i]) + ']' ]
    
    labels = ['x', 'y', 'z', 't', 'A']
    df = pd.DataFrame(f['Event_[' + str(trace_indexes[i]) + ']'], columns=labels)

    f['Event_[' + str(i) + ']'] = df.to_records(index=False)
    del f['Event_[' + str(trace_indexes[i]) + ']' ]

f.close()
```
