```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import os, sys
import tqdm
```

```python
#image_dir = '/mnt/analysis/e21072/h5test/run_0277/len90_ic600000_pads21_eps5_samps5_poly2/673876CUT_Date_12_20_2023/' # 806p
#image_dir = '/mnt/analysis/e21072/h5test/run_0277/len90_ic600000_pads21_eps5_samps5_poly2/241372CUT_Date_12_20_2023/' #1682p
#image_dir = '/mnt/analysis/e17023/Adam/GADGET2/Output/images/' # output images
image_dir = '/mnt/analysis/e17023/Adam/GADGET2/Output/aug_images/' # augmented images

image_list = list(os.listdir(image_dir))
output = []

for image in tqdm.tqdm(image_list):
    if image.endswith('png'):
        image = Image.open(image_dir+image)
        image = np.asarray(image)
        output.append(image)
```

```python
rolling_min = np.ones_like(output[0]) * 255

for image in tqdm.tqdm(output):
    rolling_min = np.minimum(rolling_min, image)

plt.figure(figsize=(6,6))
plt.imshow(rolling_min)
plt.title('Minimum Pixel Values')
```
